```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore
```

```python
df = pd.read_csv('imdb_movies.csv')
df.head()
```

```python
df.tail()
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
df = df[(np.abs(zscore(df['Rating'])) < 2.5)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Genre', y='Rating')
plt.title("Movie Ratings by Genre")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Review_Count', y='Rating', hue='Genre', s=100)
plt.title("Review Count vs. Rating")
plt.tight_layout()
plt.show()
```
